#### There are some colomns what we don't need for our analysis, so we will drop them

```python
# List of columns to drop
columns_to_drop = ['horse_power','4x4','engine_volume', 'valid_test','brand_normalized','brand_group', 'magnesium_wheels', 'distance_control', 'economical', 'adaptive_cruise_control', 'cruise_control']

# Drop the columns
cars.drop(columns=columns_to_drop, inplace=True)

# Display the updated DataFrame
cars.head()
```

### Find nulls

```python
for colomns in cars.columns:
    empty_rows = cars[colomns].isnull().sum()
    print(colomns, '=', empty_rows / len(cars) * 100, '%')
```

### Drop nulls

```python
cars_cleaned = cars[cars.isna().sum(axis=1) < 1]

rows_before = len(cars)

rows_after = len(cars_cleaned)

rows_dropped = rows_before - rows_after

rows_dropped
```

### Ratio of 2 most popular brands by engine type

```python
palette_dark_b = sns.color_palette("dark:b_r")
palette_dark_c = sns.color_palette("dark:c_r")

Hyundai = cars_cleaned[cars_cleaned['brand'] == 'Hyundai']['fuel_type'].value_counts()
Toyota = cars_cleaned[cars_cleaned['brand'] == 'Toyota']['fuel_type'].value_counts()

plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(10,5), dpi =200)

plt.bar(['hybrid','petrol','diesel','electric'], Hyundai, label='Hyunday', color=palette_dark_b, width=-0.2, align='edge', alpha=0.6)
plt.bar(['hybrid','petrol','diesel','electric'], Toyota, label='Toyota', color=palette_dark_c,width=0.2, align='edge', alpha=0.6)

plt.title('Ratio Of Two Most Popular Brands\n By Engine Type', fontsize=20, color='black')
plt.xlabel('Engine Type', color='c')
plt.ylabel('Numbers Of Cars', color='c')
plt.legend(loc='upper right')

for i,j in zip(['hybrid','petrol','diesel','electric'], Hyundai):
    label = "{:.2f}".format(j) 
    plt.annotate(label, xy=(i,j), fontsize=10, color='b', rotation=45, ha='center')
for i,j in zip(['hybrid','petrol','diesel','electric'], Toyota):
    label = "{:.2f}".format(j) 
    plt.annotate(label, xy=(i,j), fontsize=10, color='c', rotation=45)

plt.show()
```
